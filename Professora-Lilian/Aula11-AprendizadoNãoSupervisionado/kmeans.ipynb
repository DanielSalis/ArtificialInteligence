```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import sklearn
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import scale
from sklearn.preprocessing import MinMaxScaler

import sklearn.metrics as sm
from sklearn import datasets
from sklearn.metrics import confusion_matrix, classification_report

%matplotlib inline
# rcParams['figure.figsize'] = 7, 4

df = pd.read_csv("Agregation.csv")
df.head()

plt.scatter(df['labelA'], df['labelB'])

```

```python
km = KMeans(n_clusters=7, random_state=5)
km
```

```python
y_predicted = km.fit_predict(df[['labelA', 'labelB']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head(10)
```

```python
df0 = df[df.cluster==0]
df1 = df[df.cluster==1]
df2 = df[df.cluster==2]
df3 = df[df.cluster==3]
df4 = df[df.cluster==4]
df5 = df[df.cluster==5]
df6 = df[df.cluster==6]


plt.scatter(df0.labelA,df0.labelB, color='green')
plt.scatter(df1.labelA,df1.labelB, color='red')
plt.scatter(df2.labelA,df2.labelB, color='black')
plt.scatter(df3.labelA,df3.labelB, color='blue')
plt.scatter(df4.labelA,df4.labelB, color='orange')
plt.scatter(df5.labelA,df5.labelB, color='purple')
plt.scatter(df6.labelA,df6.labelB, color='grey')

plt.xlabel('Label A')
plt.ylabel('Label B')
plt.legend()
```

```python
truth = df.labelC
print(classification_report(df.labelC, df.cluster))
```

```python
figure(figsize=(15,8))
dfPathBased = pd.read_csv("PathBased.csv")
dfPathBased.head()

y = dfPathBased['labelC']
dfPathBased = dfPathBased.drop(columns=['labelC'])

plt.scatter(dfPathBased['labelA'], dfPathBased['labelB'])
```
